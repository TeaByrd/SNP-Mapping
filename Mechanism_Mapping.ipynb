```python
import pandas as pd
```

```python
snps="PATH TO THE MAPPING"
mechanisms="PATH TO NEUROMMSIG MAPPING"
```

```python
#loading snps_genes mapping
snps_genes=pd.read_csv("../MasterThesis/masterthesislab/Final New Mappings /snps_genes_concatenated.csv",delimiter=",")
#
#loading neurommsig_mechanisms
mechanisms=pd.read_csv('../MasterThesis/masterthesislab/LordickData/neurommsig_pd_full_genes.csv')
MechanismToGene=mechanisms.to_dict("list")
```

```python
#quickly remove nan values from gene lists in mechanisms mapping to enhance efficiency
def remove_float_from_list(the_list):
    return [value for value in the_list if type(value)!=float]
#
MechanismToGene_noNans={x:remove_float_from_list(y) for x,y in MechanismToGene.items()}
```

```python
print (MechanismToGene_noNans)
```

```python
#check out mechanisms
MechanismToGenetoSnp=dict()
count=0
for mechanism in MechanismToGene_noNans.keys():
    print ("Mechanism:",mechanism)
    print(MechanismToGene_noNans[mechanism])
    selected_snps=snps_genes.loc[snps_genes["GENCODE_name"].isin(MechanismToGene_noNans[mechanism])]
    genes_mapped=set(selected_snps["GENCODE_name"].tolist())
    snps_mapped=set(selected_snps["rsID"].tolist())
    MechanismToGenetoSnp[mechanism]={"All Genes":MechanismToGene_noNans[mechanism],"Genes with mapped SNPs":genes_mapped,"SNPs":snps_mapped,
                                    "SNPs-Genes Mapping":selected_snps}
    print ("Genes mapped",genes_mapped)
    print ("Alongside SNPs mapped",snps_mapped)
    print()
```

```python
df=MechanismToGenetoSnp["MAPK subgraph"]["SNPs-Genes Mapping"]
df["Mechanism"]="MAPK subgraph"

for mech in MechanismToGenetoSnp.keys():
    df_append=MechanismToGenetoSnp[mech]["SNPs-Genes Mapping"]
    df_append["Mechanism"]=mech
    df=df.append(df_append)
```
